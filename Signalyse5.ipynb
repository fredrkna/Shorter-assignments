```python
from os.path import dirname, join as pjoin
import scipy.io as sio
from scipy.io import wavfile
from matplotlib import pyplot as plt
import numpy as np
from scipy.fftpack import fft
import copy
from IPython.display import clear_output
from scipy import signal
```

```python
data_dir = pjoin(dirname(sio.__file__), 'tests', 'data')
wav_fname = pjoin(data_dir, 'guitar.wav')
```

```python
samplerate, data = sio.wavfile.read(wav_fname)
#print(f"number of channels = {data.shape[0]}")
length = data.shape[0] / samplerate
print(f"length = {length}s")
dt = 1/samplerate
```

```python
%matplotlib notebook
time = np.linspace(0., length, data.shape[0])
plt.figure(1)
plt.plot(time, data)
#plt.plot(time, data[:, 1], label="Right channel")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Sound signal from guitar")
plt.show()

#Measured signal
```

```python
startE2 = int(0.46/dt)
sluttE2 = int(1.67/dt)
tE2 = time[startE2:sluttE2]
dE2 = data[startE2:sluttE2]

NE2 = dE2.shape[0]
spE2 = np.fft.fftshift(fft(dE2))
freqE2 = np.fft.fftshift(np.fft.fftfreq(tE2.shape[-1], d=dt))

tall = np.array([1,2,3,4,5,6])
plt.figure()
plt.vlines(82.4*tall, 0, 500, "r")
plt.plot(freqE2, np.abs(spE2)/NE2) #, freq, sp.imag)
plt.xlim(0,500)
plt.title("FFT of guitar's E2 note'")
plt.xlabel("f [Hz]")
plt.ylabel("Amplitude")
```

```python
data_dir1 = pjoin(dirname(sio.__file__), 'tests', 'data')
wav_fname1 = pjoin(data_dir1, 'fysikk.wav')
```

```python
samplerate1, data1 = sio.wavfile.read(wav_fname1)
#print(f"number of channels = {data.shape[0]}")
length1 = data1.shape[0] / samplerate1
print(f"length = {length1}s")
dt1 = 1/samplerate1
```

```python
%matplotlib notebook
time1 = np.linspace(0., length1, data1.shape[0])
plt.figure(5)
plt.plot(time1, data1)
#plt.plot(time, data[:, 1], label="Right channel")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
#plt.xlim(1.8,13.8)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Sound signal from car passing by")
plt.show()

Ns = 1/dt1
print(Ns)
#Measured signal
```

```python
tIN = time1[int(8/dt1):int(9/dt1)]
dIN = data1[int(8/dt1):int(9/dt1)]

plt.figure()
NIN = dIN.shape[0]
spIN = np.fft.fftshift(fft(dIN))
#freq = np.fft.fftshift(np.fft.fftfreq(time.shape[-1]))
freqIN = np.fft.fftshift(np.fft.fftfreq(NIN, d=dt1))
plt.plot(freqIN, np.abs(spIN/NIN)) #, freq, sp.imag)
plt.title("FFT of 1st car incoming")
plt.xlabel("f [Hz]")
plt.ylabel("Amplitude")
plt.xlim(2300, 2800)
plt.ylim(0, 50)
```
