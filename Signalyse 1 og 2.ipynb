```python
import numpy as np
import scipy
from scipy import signal
from matplotlib import pyplot as plt
```

```python
t0, t1, N = -15, 15, 1000
t = np.linspace(t0, t1, N, endpoint=False)
rectangle = (0.5+0.5*signal.square(2 * np.pi *(0.05 * (t+5))))*np.heaviside(t+7.5, 1)*np.heaviside(7.5-t, 1)
plt.xlabel("Time [ms]")
plt.ylabel("Amplitude")
plt.title("A rectangular pulse")
plt.plot(t, rectangle)
```

```python
t = np.linspace(t0, t1, N)
triangle = 0.5*(1+signal.sawtooth(0.267 * np.pi * t))*np.heaviside(t, 1)*np.heaviside(7.5-t, 1)
plt.plot(t,triangle)
plt.title("Triangular pulse")
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
```

```python
t = np.linspace(t0, t1, N, endpoint=False)
i, q, e = signal.gausspulse(t, fc=1/np.e, bw = 0.5, retquad=True, retenv=True)
plt.plot(t, e)
plt.title("Gaussian pulse")
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
```

```python
t = np.linspace(t0, t1, N, endpoint=False)
plt.figure()
sig = ((t+1.5)%10) < 3
plt.plot(t, sig)
plt.title("Rectangular pulse sequence")
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
```

```python
conv1 = np.convolve(rectangle, rectangle, "same")/sum(rectangle)
#tc = np.linspace(t0, t1, 2*N-1)
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
plt.title("Convolution of rectangle with itself")
plt.plot(t, conv1)
```

```python
conv2 = signal.convolve(rectangle, triangle, "same")/sum(triangle)
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
plt.title("Convolution of rectangle with triangle")
plt.plot(t, conv2)
```

```python
conv3 = signal.convolve(e, sig, "same")/sum(sig)
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
plt.title("Convolution of gaussian pulse with pulse sequence")
plt.plot(t, conv3)
```

```python
conv3 = signal.convolve(triangle, sig, "same")/sum(sig)
plt.ylabel("Amplitude")
plt.xlabel("Time [ms]")
plt.title("Convolution of triangle with pulse sequence")
plt.plot(t, conv3)
```
