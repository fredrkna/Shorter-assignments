```python
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0, 1, 1000)
u = np.heaviside(t, 1)
x0 = np.array([1, 0])

A = np.array([[-6, -5], [1, 0]])
B = np.array([[0], [2]])
C = np.array([[1, 1]])
D = np.array([[2]])

sys = signal.lti(A, B, C, D)
ts, ys = signal.step2(sys, X0 = x0, T = t)
plt.title("Step response")
plt.xlabel("t [s]")
plt.ylabel("Amplitude")
plt.plot(ts, ys)
```
