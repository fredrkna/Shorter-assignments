```python
import numpy as np
from matplotlib import pyplot as plt
import copy
```

```python
dx = np.linspace(0, np.pi, 1000)
r = 0.5
FTCS = 1 - 4*r*np.sin(dx/2)**2
BTCS = 1/(1+4*r*np.sin(dx/2)**2)
Excact = np.exp(-1*r*dx**2)
plt.plot(dx, FTCS, label = "FTCS")
plt.plot(dx, BTCS, label = "BTCS")
plt.plot(dx, Excact, label = "Excact")
plt.title("r = 1/2")
plt.xlabel("k*deltax")
plt.ylabel("G")
plt.legend()
```

```python
r = 1/6
FTCS = 1 - 4*r*np.sin(dx/2)**2
BTCS = 1/(1+4*r*np.sin(dx/2)**2)
Excact = np.exp(-1*r*dx**2)
plt.plot(dx, FTCS, label = "FTCS")
plt.plot(dx, BTCS, label = "BTCS")
plt.plot(dx, Excact, label = "Excact")
plt.title("r = 1/6")
plt.xlabel("k*deltax")
plt.ylabel("G")
plt.legend()
```

```python
#BTCS's amplification factor decreases at a slower rate than FTCS, which I interpret as BTCS converging faster than FTCS.
```

```python
def FTCS(tEnd, jmax, r):
    dy=1/(jmax-1)
    u = np.zeros(jmax)
    u[-1] = 1
    y = np.linspace(0,1,jmax)
    dt = r*dy**2
    nmax = int(np.ceil(tEnd/dt))
    for n in range(nmax):
        nu = copy.copy(u*(1-2*r))
        nu[1:] += copy.copy(r*u[:-1])
        nu[:-1] += copy.copy(r*u[1:])
        nu[0]=0
        nu[-1]=1
        u = copy.copy(nu)
    return y, u

y2, u2 = FTCS(0.1, 101, 0.5001)
y3, u3 = FTCS(0.1, 101, 0.501)
y4, u4 = FTCS(0.1, 101, 0.51)

plt.figure(1)
plt.title("r = 0.5001")
plt.xlabel('u(y)')
plt.ylabel('y')
plt.plot(y2,u2)

plt.figure(2)
plt.title("r = 0.501")
plt.xlabel('u(y)')
plt.ylabel('y')
plt.plot(y3,u3)

plt.figure(3)
plt.title("r = 0.51")
plt.xlabel('u(y)')
plt.ylabel('y')
plt.plot(y4,u4)
```
